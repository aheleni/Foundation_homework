```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
rows = []
female_authors = doc.find_all('tr')
for author in female_authors:
    row = {}
    row['title'] = author.find('a', class_='bookTitle').get_text(strip=True)
    row['rank'] = author.find('td', class_='number').get_text(strip=True)
    row['author'] = author.find('span', itemprop='name').get_text(strip=True)
    row['score'] = author.find('span', class_='smallText uitext').find_all('a')[0].text
    row['votes'] = author.find('span', class_='smallText uitext').find_all('a')[1].text
    row['rating'] = author.find('span', class_='minirating').get_text(strip=True)
    
    rows.append(row)
```
